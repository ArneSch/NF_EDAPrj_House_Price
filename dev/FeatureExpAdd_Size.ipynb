```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_raw = pd.read_csv('King_County_House_prices_dataset.csv')
df_raw['qm_living'] = df_raw.sqft_living*0.092903
df_raw['qm_lot'] = df_raw.sqft_lot*0.092903
```

```python
df_raw.sqft_living.hist(bins=40)
plt.show()
df_raw.sqft_lot.hist(bins=40)
plt.show()
```

#### lot size outlyer or the avarge range

```python
df_raw.qm_lot.plot()
df_avrsized = df_raw[df_raw.qm_lot < df_raw.qm_lot.mean() + df_raw.qm_lot.std()]
df_avrsized.qm_lot.plot()
print(df_raw.shape,df_avrsized.shape)
```

```python
df_avrsized.qm_living.plot()
df_avrsized2 = df_avrsized[df_avrsized.qm_living < df_avrsized.qm_living.mean() + df_avrsized2.qm_living.std()]
df_avrsized2.qm_living.plot()
print(df_avrsized.shape,df_avrsized.shape)
```

```python
sns.jointplot(x='qm_living',y='price',data=df_avrsized2,kind='scatter')
sns.jointplot(x='qm_lot',y='price',data=df_avrsized2,kind='hex')
```

```python
## box in the lot size by 2000qm
df_avrsized2[df_avrsized2.qm_lot< 2000].shape
```

### lot price seemst to have two dir



```python
gz = df_avrsized2.groupby('zipcode')
plt.figure(figsize=(12,5))
gz.median()['qm_lot'].plot.bar()
plt.show()
plt.figure(figsize=(12,5))
gz.max()['qm_lot'].plot.bar()
plt.show()
```
