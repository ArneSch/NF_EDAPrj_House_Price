```python
# https://www.kingcounty.gov/
# https://en.wikipedia.org/wiki/King_County,_Washington
```

```python
us_cities[us_cities.state_id == 'WA'].county_name.unique()
```

```python
us_cities[(us_cities.state_id == 'WA') & (us_cities.county_name == 'King')].city.unique()
```

```python
c = us_cities[(us_cities.state_id == 'WA') & (us_cities.county_name == 'King')]
```

```python
cns = c[(c.city != 'Seattle') & (c.population > 20000)]
plt.figure(figsize=(12,5))
plt.bar(cns.city,cns.population)
plt.xticks(cns.city, rotation='vertical', size=8)
plt.show()
cns.shape
```
