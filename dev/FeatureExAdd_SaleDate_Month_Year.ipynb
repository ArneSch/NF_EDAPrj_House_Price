```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_raw = pd.read_csv('King_County_House_prices_dataset.csv')
```

### Sale Time range Explore

#### Get the year

```python
import re
temp = df_raw['date']
temp = temp.apply(lambda x: int(re.findall('\d+',x)[2]))
df_raw['sale_year'] = temp
#df_raw.head()
#df_raw.info()
#df_raw.plot.scatter(x='sale_year',y='price')
df_raw.sale_year.plot.hist()
```

#### Get the Month - actual time 

```python
import re
temp = df_raw['date']
temp = temp.apply(lambda x: int(re.findall('\d+',x)[0]))
df_raw['sale_month'] = temp
#df_raw.head()
#df_raw.info()
#df_raw.plot.scatter(x='sale_month',y='price')
df_raw.sale_month.plot.hist()
```

#### Check the dist. over the year and the time range

```python
for n,y in df_raw.groupby('sale_year'):
    y['sale_month'].plot.hist(bins=range(1,13))
    plt.title(n)
    plt.show()
```
