```python
import shapefile as shp  # Requires the pyshp package
import matplotlib.pyplot as plt

sf = shp.Reader("../data/Zipcodes_for_King_County_and_Surrounding_Area__zipcode_area.shp")

plt.figure()
for shape in sf.shapeRecords():
    x = [i[0] for i in shape.shape.points[:]]
    y = [i[1] for i in shape.shape.points[:]]
    plt.plot(x,y,'gray')
plt.show()
```

```python
import shapefile as shp  # Requires the pyshp package
import matplotlib.pyplot as plt

sf = shp.Reader("../data/Street_Directional_Zones__County_and_CitySpecific__street_direction_zones_area.shp")

plt.figure()
g = lambda x: (min(x), max(x))
for i,shape in enumerate(sf.shapeRecords()):
    if i == 300:
        break
    x = [i[0] for i in shape.shape.points[:]]
    
    y = [i[1] for i in shape.shape.points[:]]
    plt.plot(g(x),g(y),marker='o',color='gray')
plt.show()
```
